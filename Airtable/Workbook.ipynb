```julia
import Pkg; Pkg.add("JSONTables")
using CSV, DataFrames, DataFramesMeta, Dates, JSONTables, Glob, Plots, Random
```

```julia
csv_files = glob("*/*.csv")
```

```julia
K = DataFrame.(CSV.File.(csv_files))
```

```julia
K_aggregate = reduce(vcat, K)
```

```julia
K_label = groupby(K_aggregate, :Label)
```

```julia
Kiwi = K_label[2]
```

```julia
select!(Kiwi, Not([:Image, :Audio, :Label, :Comment, :Collaborators]))
```

```julia
sort!(Kiwi, :"File Name")
```

```julia
@transform!(Kiwi, @byrow :"Location" = (split(:"Trap Number", "/"))[end] )
```

```julia
@transform!(Kiwi, @byrow :"File" = (split(:"File Name", "-"))[2] )
```

```julia
@transform!(Kiwi, @byrow :"S" = (split(:"File Name", "-"))[3][2:end] )
@transform!(Kiwi, @byrow :"E" = (split(:"File Name", "-"))[4][2:end] )
```

```julia
describe(Kiwi)
```

```julia
select!(Kiwi, Not([:"File Name", :"Trap Number"]))
```

```julia
Kiwi_grouped = groupby(Kiwi, :File)
```

```julia
for group in Kiwi_grouped
	print("[")
	print("""{Reviewer:"D", Operator:"Airtable", Duration:895},""")
	for row in eachrow(group)
	    print("""[$(row.S), $(row.E), 0, 8000, [{species:"K-Set", calltype:"$(row."Call Type")", "filter": "Kiwi (Tokoeka Fiordland)", certainty:99}]],""")
	end
	print("]\n\n")
end;
```
