```julia
Kiwi=DataFrame(CSV.File("Kiwi.csv"))
```

```julia
sort!(Kiwi, :"File Name")
```

```julia
@transform!(Kiwi, @byrow :"File" = (split(:"File Name", "-"))[2] )
@transform!(Kiwi, @byrow :"S" = (split(:"File Name", "-"))[3][2:end] )
@transform!(Kiwi, @byrow :"E" = (split(:"File Name", "-"))[4][2:end] )
```

```julia
describe(Kiwi)
```

```julia
levels(Kiwi."Trap Number")
```

```julia
@transform!(Kiwi, @byrow :"Trap Number" = (split(:"Trap Number", "/"))[2] )
```

```julia
levels(Kiwi."Trap Number")
```

```julia
Kiwi_grouped = groupby(Kiwi, :"Trap Number")
```

```julia
p=sort((combine(Kiwi_grouped, nrow)), :nrow)
```

```julia
paths=p."Trap Number"
```

```julia
mkpath.(paths)
```

```julia
for loc in Kiwi_grouped
	B = groupby(loc, :File)
	for group in B
		io = open("$(loc[1,5])/$(group[1,6]).wav.data", "w")
		write(io, """[ """)
		write(io, """{"Reviewer":"D", "Operator":"Airtable", "Duration":895},""")
		for row in eachrow(group)
			write(io, """[$(row.S), $(row.E), 0, 8000, [{"species":"K-Set", "calltype":"$(row."Call Type"[1])", "filter": "Kiwi (Tokoeka Fiordland)", "certainty":99}]]""")
			if row != last(group) 
				write(io, """,""")
			end
		end
		write(io, """ ]""")
		close(io)
	end
end
```
